# Linear Regression


```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("Hours and Scores.csv")

plt.scatter(data.Hours, data.Scores)

plt.show()
```

# Cost function

```python
def loss_function(w, b, N : int):
    """ Calculates the cost function for the dataset"""
    total_error = 0
    for i in range(len(N)):
        X = N.iloc[i].Hours
        y = N.iloc[i].Scores
        total_error += ( y - (w * X +b)) ** 2
    return total_error / float(len(N))    
   
```

```python
print(loss_function(9.984794012292527, 1.576849619962064, data))
```

# Gradient Descent

```python
def Gradient_Descent(w_now, b_now, N, alpha):
    w_gradient = 0
    b_gradient = 0
    
    n = len(N)
    for i in range(n):
        X = N.iloc[i].Hours
        y = N.iloc[i].Scores
        
        w_gradient += -(2/n) * X * (y- (w_now * X + b_now))
        b_gradient += -(2/n) * (y - (w_now * X + b_now))
    w = w_now - w_gradient * alpha
    b = b_now - b_gradient * alpha
    
    return w, b

w = 0
b =0
l = 0.0001
epochs = 1000

for i in range(epochs):
    w, b = Gradient_Descent(w, b, data, l)
    print(f"for {i} iteration the w : {w}, b: {b}")
print(w, b)    

plt.scatter(data.Hours,data.Scores, color = "black")
plt.plot(list(range(1, 10)), [w * x + b for x in range(1, 10)], color='red')
plt.show()
        
            
    
```
